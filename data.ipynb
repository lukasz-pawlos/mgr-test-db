```python
import os
import pandas as pd

def sort():
    # Get the current working directory where the script is located
    current_dir = os.getcwd()
    
    # Iterate through the directories test1, test2, test3
    for test_dir in ['test1', 'test2', 'test3']:
        test_dir_path = os.path.join(current_dir, test_dir)
        
        # Walk through all subdirectories in test1, test2, test3
        for root, dirs, files in os.walk(test_dir_path):
            for file in files:
                if file.endswith('.csv'):
                    file_path = os.path.join(root, file)
                    
                    # Read the CSV file
                    df = pd.read_csv(file_path)
                    
                    # Sort the DataFrame by the 'timestamp' column
                    df = df.sort_values(by='timestamp')
                    
                    # Save the sorted DataFrame back to the CSV file
                    df.to_csv(file_path, index=False)
                    print(f'Sorted and saved: {file_path}')

# Run the function
sort()
```

```python
def process_logs():
    # Pobieramy bieżący katalog, w którym znajduje się skrypt
    current_dir = os.getcwd()
    
    # Iterujemy przez katalogi test1, test2, test3
    for test_dir in ['test1', 'test2', 'test3']:
        test_dir_path = os.path.join(current_dir, test_dir)
        
        # Przechodzimy przez wszystkie podkatalogi w test1, test2, test3
        for root, dirs, files in os.walk(test_dir_path):
            # Sprawdzamy, czy w podkatalogu znajdują się pliki traffic_log.csv i market_log.csv
            if 'traffic_log.csv' in files and 'market_log.csv' in files:
                traffic_log_path = os.path.join(root, 'traffic_log.csv')
                market_log_path = os.path.join(root, 'market_log.csv')
                
                # Wczytujemy pliki CSV do DataFrame
                traffic_log_df = pd.read_csv(traffic_log_path)
                market_log_df = pd.read_csv(market_log_path)
                
                # Łączymy dane po kolumnie requestId
                complete_log_df = pd.merge(market_log_df, traffic_log_df[['requestId', 'apiTime']], on='requestId', how='inner')
                
                # Wybieramy tylko potrzebne kolumny do finalnego pliku
                complete_log_df = complete_log_df[['id', 'timestamp', 'apiMethod', 'applicationTime', 'databaseTime', 'endpointUrl', 'apiTime']]
                
                # Zapisujemy wynik do nowego pliku CSV w tym samym podkatalogu
                output_path = os.path.join(root, 'complete_market_log_csv.csv')
                complete_log_df.to_csv(output_path, index=False)
                print(f"Plik zapisany w: {output_path}")

# Uruchamiamy funkcję przetwarzającą logi
process_logs()
```

```python
import re

def filter_records():
    # Pobieramy bieżący katalog, w którym znajduje się skrypt
    current_dir = os.getcwd()
    
    # Iterujemy przez katalogi test1, test2, test3
    for test_dir in ['test1', 'test2', 'test3']:
        test_dir_path = os.path.join(current_dir, test_dir)
        
        # Przechodzimy przez wszystkie podkatalogi
        for root, dirs, files in os.walk(test_dir_path):
            # Sprawdzamy, czy w podkatalogu znajduje się plik complete_market_log_csv.csv
            if 'complete_market_log_csv.csv' in files:
                file_path = os.path.join(root, 'complete_market_log_csv.csv')
                
                # Wczytujemy plik CSV
                df = pd.read_csv(file_path)
                
                # Definiujemy wzorce dla endpointUrl
                allowed_patterns = [
                    r"^/buyoffer/create$",
                    r"^/selloffer/create$",
                    r"^/stockrate/company/\d+$"  # Dopuszcza dynamiczne ID w formie liczbowej
                ]
                
                # Łączymy wzorce w jedno wyrażenie regularne
                combined_pattern = re.compile('|'.join(allowed_patterns))
                
                # Filtrujemy rekordy za pomocą wyrażeń regularnych
                filtered_df = df[df['endpointUrl'].apply(lambda x: bool(combined_pattern.match(x)))]
                
                # Nadpisujemy plik CSV przefiltrowanymi danymi
                filtered_df.to_csv(file_path, index=False)
                print(f"Przefiltrowany plik zapisany w: {file_path}")

filter_records()
```

```python
import os
import pandas as pd

def process_trade_logs():
    # Pobieramy bieżący katalog, w którym znajduje się skrypt
    current_dir = os.getcwd()
    
    # Iterujemy przez katalogi test1, test2, test3
    for test_dir in ['test1', 'test2', 'test3']:
        test_dir_path = os.path.join(current_dir, test_dir)
        
        # Przechodzimy przez wszystkie podkatalogi w test1, test2, test3
        for root, dirs, files in os.walk(test_dir_path):
            # Sprawdzamy, czy w podkatalogu znajduje się plik trade_log.csv
            if 'trade_log.csv' in files:
                file_path = os.path.join(root, 'trade_log.csv')
                
                # Wczytujemy plik trade_log.csv
                df = pd.read_csv(file_path)
                
                # Wyświetlamy dostępne kolumny, aby upewnić się, że kolumna 'timestump' istnieje
                print(f"Kolumny w pliku {file_path}: {df.columns.tolist()}")
                
                # Sprawdzamy, czy kolumna 'timestump' istnieje
                if 'timestamp' in df.columns:
                    # Konwersja kolumny 'timestump' na datetime, jeśli jeszcze nie jest w tym formacie
                    df['timestamp'] = pd.to_datetime(df['timestamp'], format='ISO8601')
                    
                    # Sortowanie danych po kolumnie 'timestump'
                    df = df.sort_values(by='timestamp')
                    
                    # Sumowanie kumulatywne kolumn numberOfSellOffers i numberOfBuyOffers
                    df['numberOfSellOffers'] = df['numberOfSellOffers'].cumsum()
                    df['numberOfBuyOffers'] = df['numberOfBuyOffers'].cumsum()
                    
                    # Zapisujemy wynik do nowego pliku CSV w tym samym podkatalogu
                    output_path = os.path.join(root, 'sum_trade_log.csv')
                    df.to_csv(output_path, index=False)
                    print(f"Plik sum_trade_log.csv zapisany w: {output_path}")
                else:
                    print(f"Kolumna 'timestamp' nie została znaleziona w pliku {file_path}.")

# Uruchamiamy funkcję przetwarzającą pliki trade_log.csv
process_trade_logs() 
```
